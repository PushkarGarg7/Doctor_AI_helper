```python
import random
from reportlab.lib.pagesizes import letter
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Image, Table, TableStyle
from reportlab.lib.styles import getSampleStyleSheet

# Data to be added
name = "Abhinav Aggarwal"
age = 22
gender = "Male"
image_path = "C:\Abhinav\Abhinav\PEC\Major Project\Project\Doctor_AI_helper\img1.png"  # Path to your image
text_data = """
This is an example of some detailed information that will appear in the PDF. 
You can add more lines to make it 3-6 lines long.
"""

# List of diseases with random probabilities
diseases = [
    "Atelectasis", "Consolidation", "Infiltration", "Pneumothorax", "Edema", 
    "Emphysema", "Fibrosis", "Effusion", "Pneumonia", "Pleural_thickening", 
    "Cardiomegaly", "Nodule Mass", "Hernia"
]
disease_probabilities = [(disease, round(random.uniform(0, 1), 2)) for disease in diseases]

# Create a PDF document
pdf = SimpleDocTemplate("preliminary_report.pdf", pagesize=letter)

# Set styles
styles = getSampleStyleSheet()
title_style = styles['Title']
normal_style = styles['Normal']
heading_style = styles['Heading1']

# Components for the PDF
title = Paragraph("Preliminary Report", heading_style)
image_heading = Paragraph("X-Ray Image", heading_style)

# Disease Likelihood Table
disease_table_data = [["Disease", "Probability"]] + [[disease, f"{prob*100}%"] for disease, prob in disease_probabilities]
disease_table = Table(disease_table_data, colWidths=[200, 100])
disease_table.setStyle(TableStyle([
    ('BACKGROUND', (0, 0), (-1, 0), colors.grey),
    ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
    ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
    ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
    ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
    ('BACKGROUND', (0, 1), (-1, -1), colors.beige),
    ('GRID', (0, 0), (-1, -1), 1, colors.black),
]))

# Image and Text
image = Image(image_path, width=150, height=150)
text_paragraph = Paragraph(text_data, normal_style)

# Create the PDF layout
elements = [
    title, Spacer(1, 12),
    image_heading, Spacer(1, 6),
    image, Spacer(1, 20),
    disease_table, Spacer(1, 20),
    text_paragraph
]

# Build the PDF
pdf.build(elements)

print("PDF created successfully!")
```
